```python
#POBLACIÃN TOTAL 
PT =  df_cgsse['FEX_C18'].sum()
print(f' La poblaciÃ³n total de qeureferencia al pais colombia Colombia es de {PT} personas')
```

```python
#POBLACION EN EDAD DE TRABAJAR 
PET = df_cgsse[(df_cgsse['P6040']>=15)]['FEX_C18'].sum()
print(f" La poblaciÃ³n en edad de trabajar (mayor o iguales a 15 aÃ±os) para Colombia es de {PET} ")
```

```python
#POBLACIÃN ECONOMICAMENTE ACTIVA
PEA=df_ft[(df_ft['FT']==1)]['FEX_C18'].sum()
print(f'PoblaciÃ³n Economicamente Activa (PEA) para colombia es de {PEA} personas')
```

```python
#TASA DE DESEMPLEO
D=df_desocupados[(df_desocupados['DSI']==1)]['FEX_C18'].sum()
print(f'la PoblaciÃ³n Desocupada para colombia es de {D} personas')
TD = D/PEA
print(f'la Tasa de Desempleo (TD), calculada como la poblacion desocupada(D), dividida entre la poblacion economicament activa (PEA) es de: {TD:.2%}')
```
